```python
import pandas as pd

# 1. CSV 불러오기
df = pd.read_csv('../births.csv')

# 2. day 결측치 제거 및 정수화
df = df.dropna(subset=['day'])
df['day'] = df['day'].astype(int)

# 3. 문자열 조합 → 날짜 파싱 (format 적용 + 오류 무시)
df['ymd'] = (
    df['year'].astype(str) +
    df['month'].astype(str).str.zfill(2) +
    df['day'].astype(str).str.zfill(2)
)

df['date'] = pd.to_datetime(df['ymd'], format='%Y%m%d', errors='coerce')

# 4. datetime → Unix timestamp 변환 (나노초 → 밀리초 → *1000)
df['timestamp_ms'] = df['date'].view('int64') // 10**6
df['timestamp_ms'] = df['timestamp_ms'] * 1000  # ✔️ 여기 *1000

# 5. 결과 확인
print(df[['date', 'timestamp_ms']].head())
```

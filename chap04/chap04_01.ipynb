```python
import pandas as pd
import numpy as np
```

```python
def make_df(cols, ind):
    "빠르게 DataFrame 생성"
    data = {c: [str(c)+str(i) for i in ind]
        for c in cols}
    return pd.DataFrame(data, ind)

```

```python
df_area = pd.read_csv('state-areas.csv')
df_area.head(), df_area.shape
```

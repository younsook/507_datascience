```python
import matplotlib.pyplot as plt
from itertools import cycle

colors = cycle(['red', 'green', 'blue'])
x = [1, 2, 3, 4, 5]
for i in range(3):
    y = [j * (i + 1) for j in x]
    plt.plot(x, y, color=next(colors), label=f'{i+1}배')

plt.legend()
plt.title("cycle과 matplotlib 예시")
plt.show()
```

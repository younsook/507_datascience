```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Iris 데이터셋 로드
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = [iris.target_names[i] for i in iris.target]

df.head()
```

```python
sns.boxplot(data=df,
            x="species",
            y="sepal length (cm)",
            palette="pastel")
plt.title("Sepal Length Distribution by Species")
plt.show()
```

```python
sns.pairplot(df, hue="species", palette="Set2")
plt.suptitle("Iris Feature Pairwise Plot", y=1.02)
plt.show()
```

```python
mean_values = df.groupby("species")["petal width (cm)"].mean().reset_index()

sns.barplot(data=mean_values,
            x="species",
            y="petal width (cm)",
            palette="muted")
plt.title("Average Petal Width by Species")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# 데이터 로딩
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# 시각화
sns.pairplot(df, hue='species')  # species 기준으로 색상 다르게
plt.show()
```

## seaborn에서 pairplot() 또는 PairGrid로 자동으로 해주는 기능

hue='species'	클래스별 색상 구분 (target 값을 기준으로 색칠)
diag_kind='hist'	대각선에 히스토그램 (kde도 가능)
kind='scatter' (기본)	비대각선은 산점도
corner=True	하삼각형만 출력

```python
import seaborn as sns

sns.pairplot(df, hue="species", diag_kind="hist")
plt.show()
```

```python
#기본적으로 시각화할때 색 사용하지 말기
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x - 0), color ="black")  # 컬러에 이름이 있는것은 써도됨.
plt.plot(x, np.cos(x - 1), color ="g")  # cos(x - 1)
plt.plot(x, np.sin(x - 2), color ="0.75")  # 회색의 그라데이션
plt.plot(x, np.cos(x - 3), color ="#FFDD44")  # RGB색은 사용하지말것.
plt.plot(x, np.sin(x - 4), color =(1.0, 0.2, 0.3))  # sin(x - 4)
plt.plot(x, np.cos(x - 5), color ="chartreuse")  # cos(x - 5)
```

```python
#기본적으로 시각화할때 색 사용하지 말기
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x - 0), linestyle="solid")
plt.plot(x, np.cos(x - 1), linestyle="dashed")
plt.plot(x, np.sin(x - 2), linestyle="dashdot")
plt.plot(x, np.cos(x - 3), linestyle="dotted")
plt.plot(x, np.sin(x - 4), )
plt.plot(x, np.cos(x - 5), )
```

```python
# 데이터 로드
iris = load_iris()
x = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

df = pd.DataFrame(x, columns=feature_names)
df["species"] = [target_names[i] for i in y]

df.tail(3)
```

```python
#petal length (cm )가 1.5보다 큰 데이터를 출력
df[df["petal length (cm)"] > 1.5] #boolean data frame -> 필터링
```

```python
#x,y축 뽑아서 라벨링 먼저하고 의미있는데이터를 찾는다.
df_petal = df[df["petal length (cm)"] > 1.5]
ax = df_petal.plot()
ax.set_title("Iirs 데이터")
ax.set_xlabel("개별 데이터")
ax.set_ylabel("관측 데이터")
```

```python
# 데이터 로드
iris = load_iris()
x = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names
# DataFrame 생성
df = pd.DataFrame(x, columns=feature_names)
df["species"] = [target_names[i] for i in y]

# 결과 출력
df.tail(3)
```

```python
means = df.groupby("species").mean()

colors = ["red", "blue", "yellow"]
species_list = iris.target_names
```

```python
#means.head()
```

```python
#plt.plot("sepal length (cm)", means.loc["setosa"])
plt.figure(figsize=(12,8))
for i, species in enumerate(species_list):
    #print(i, species)
    plt.plot(iris.feature_names, means.loc[species]
             , color=colors[i], label=species)

plt.title("품종별 특성 평균값 비교")
plt.xlabel("특성")                       
plt.ylabel("평균값 (cm)")  
plt.show
```
